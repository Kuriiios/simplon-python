### 1

```python
lotr = "Three Rings for the Elven-kings under the sky, Seven for the Dwarf-lords in their halls of stone, Nine for Mortal Men doomed to die, One for the Dark Lord on his dark throne In the Land of Mordor where the Shadows lie. One Ring to rule them all, One Ring to ï¬nd them, One Ring to bring them all and in the darkness bind them In the Land of Mordor where the Shadows lie."
```

```python
lotr = lotr.replace('.', '').replace(',', '').lower()
list_words = lotr.split(' ')
word = rd.choice(list_words)
hidden_word = len(word) * '_'
```

### 2

```python
n_try =  8
guess = ''

list_hidden_word = list(hidden_word)
chars = list(word)
vocab_size = len(chars)

stoi = {ch : i for i, ch in enumerate(chars)}
itos = {i : ch for ch, i in stoi.items()}

while (guess != word and n_try > 0):
    wrong = True
    guess = input("Choose a character :")
    for i, ch in enumerate(word):
        if guess == ch:
            list_hidden_word[i] = word[i]
            hidden_word = ''.join(list_hidden_word)
            wrong = False
    if wrong == True:
        n_try -=1
        print(f"You have {n_try} left : {hidden_word}")
    else:
        print(hidden_word)
    if word == hidden_word:
        guess = word
        print("you won")
    if n_try == 0:
        print(f"you loser, the word was {word}")
```
