```python
import string
import random
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output
```

### 1

```python
fig_radius = 5
def afficher_pendu(nb_erreurs):
    """
    Affiche le pendu en fonction du nombre d'erreurs.
    """


    plt.figure(figsize=(fig_radius, fig_radius))  # Crée une nouvelle figure

    # Dessine la potence
    plt.plot([0, 1], [0, 0], 'k')  # Base
    plt.plot([0.5, 0.5], [0, 1], 'k')  # Poteau vertical
    plt.plot([0.5, 1], [1, 1], 'k')  # Barre horizontale
    plt.plot([1, 1], [1, 0.8], 'k')  # Corde

    # Dessine le corps du pendu en fonction du nombre d'erreurs
    if nb_erreurs > 0:# Tête
        plt.plot(1, 0.7 , marker="o",  markersize=fig_radius*10, color='pink', mec='black')

    if nb_erreurs > 1:# Corps
        plt.plot([1, 1], [0.3, 0.55], 'k')
    if nb_erreurs > 2:# Bras gauche
        plt.plot([0.8, 1], [0.6, 0.4], 'k')
    if nb_erreurs > 3:# Bras droitw
        plt.plot([1, 1.2], [0.4, 0.6], 'k')
    if nb_erreurs > 4:# Jambe gauche
        plt.plot([1, 0.8], [0.3, 0.1], 'k')
    if nb_erreurs > 5:# Jambe droite
        plt.plot([1, 1.2], [0.3, 0.1], 'k')
    if nb_erreurs > 6:# End
        plt.clf()
        plt.plot(0.6, 0.5 , marker="o",  markersize=fig_radius*40, color='pink', mec='black')
        plt.plot( 0.4, 0.6 , 50, marker="o", color='black') #left eye
        plt.plot( 0.8, 0.6 , 50, marker="o", color='black') #right eye
        plt.plot([0.6, 0.3], [0.4, 0.2], 'k') #mouth left
        plt.plot([0.6, 0.9], [0.4, 0.2], 'k') #mouth right

    # Paramètres d'affichage
    plt.axis('off')  # Cache les axes
    plt.xlim(-0.2, 1.4)  # Ajuste les limites de l'axe x
    plt.ylim(-0.2, 1.4)  # Ajuste les limites de l'axe y
    plt.show()  # Affiche le pendu
```

```python
lotr = "Three Rings for the Elven-kings under the sky, Seven for the Dwarf-lords in their halls of stone, Nine for Mortal Men doomed to die, One for the Dark Lord on his dark throne In the Land of Mordor where the Shadows lie. One Ring to rule them all, One Ring to ﬁnd them, One Ring to bring them all and in the darkness bind them In the Land of Mordor where the Shadows lie."
```

```python
alphabet = list(string.ascii_lowercase)
lotr = lotr.replace('.', '').replace(',', '').replace('-', '').lower()
list_words = lotr.split(' ')
word = random.choice(list_words)
hidden_word = len(word) * '_'
n_try =  8
guess = ''
print(f'The word is {len(hidden_word)} long :')
```

### 2

```python
while (guess != word and n_try > 0):
    guess = input("Choose a character :").lower()
    while guess not in alphabet and len(guess) != 1:
        guess = input("Unused letter only :").lower()
        alphabet.remove(guess)
    for i, ch in enumerate(word):
        if ch == guess:
            hidden_word = hidden_word[:i] + word[i] + hidden_word[i+1:]
    if guess not in word :
        n_try -=1
        clear_output(wait=False)
        afficher_pendu(7 - n_try)
        print(f"You have {n_try} tries left : ")
    print(hidden_word)
    if word == hidden_word:
        guess = word
        print(f'you won with {8 - n_try} errors the word was : {word}')
    if n_try == 0:
        print(f'try again later, the word was: {word}')
```
