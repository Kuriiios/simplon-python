```python
import random
```

```python
possibilities = ['rock', 'paper', 'scissors']
computer_choice = random.choice(possibilities)
```

```python
humain_choice = input("Choose between rock, paper and scissors")
match computer_choice:
    case 'rock':
        match humain_choice:
            case 'rock':
                print('Draw')
            case 'paper':
                print('You won')
            case 'scissors':
                print('Computer won')
    case 'paper':
        match humain_choice:
            case 'rock':
                print('Computer won')
            case 'paper':
                print('Draw')
            case 'scissors':
                print('You won')
    case 'scissors':
        match humain_choice:
            case 'rock':
                print('You won')
            case 'paper':
                print('Computer won')
            case 'scissors':
                print('Draw')
```
