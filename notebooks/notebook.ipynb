```python
PATH  = r'.venv' 
Nroots = sum(len(root) for root, dirnames, filesnames in os.walk(PATH) )
Ndirs = sum(len(dirnames) for root, dirnames, filesnames in os.walk(PATH) )
Nfiles = sum(len(filesnames) for root, dirnames, filesnames in os.walk(PATH) ) 
print(f"Number of roots: {Nroots}, Number of files: {Nfiles}, Number of directories: {Ndirs}")
```

```python
from matplotlib import pyplot as plt
import numpy as np 
```

```python
start = 0
stop = 9
num = 20
t = np.linspace(start, stop, num)
t
```

```python
fig = plt.figure(figsize=(4,4))

a = 1
b = 0
c = 1.5
d= -1

funct= a*t +b
funct2= c*t +d

plt.plot(t, funct, "--", color = 'orange', lw=3)
plt.plot(t, funct2, "--", color = 'purple', lw=3)
plt.grid()
plt.xlim(t.min(), t.max())
plt.ylim(np.minimum(funct.min(), funct2.min()), np.maximum(funct.max(), funct2.max()))
```
