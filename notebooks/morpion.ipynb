```python
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import clear_output
```

```python
arr = np.full((3,3), 2)
stoi = {0:'o', 1:'X', 2:None}
```

```python
def playing():
    x = -1
    y = -1
    while x not in range(0, 3):
        try:
            x = int(input('X value in array'))
        except ValueError:
            print("Should be an int not a character")
    while y not in range(0, 3):
        try:
            y = int(input('Y value in array'))
        except ValueError:
            print("Should be an int not a character")
    return x, y

def show_gril(arr):
    plt.plot([0.5, 0.5], [2.5, -0.5], 'k')
    plt.plot([1.5, 1.5], [2.5, -0.5], 'k')
    plt.plot([-0.5, 2.5], [0.5, 0.5], 'k')
    plt.plot([-0.5, 2.5], [1.5, 1.5], 'k')
    plt.xticks(range(3))
    plt.yticks(range(3))
    arr = arr.reshape(9)
    for i in range(9):
        plt.plot( i//3, i%3 , marker=stoi[arr[i]], markerfacecolor='white', markersize=30,  color='black')
    return plt.show()
```

```python
def morpion():
    finished = False
    i = 0
    while finished == False:
        x, y = playing()
        clear_output(wait=False)
        mask = arr[x, y] == 0
        if mask == True:
            print('combination not possible, play again')
            pass
        else :
            i += 1
            arr[x, y] = (i % 2)
            if np.array_equal(np.full((3), i % 2), arr[x]) or np.array_equal(np.full((3), i % 2), np.array((arr[0][y], arr[1][y], arr[2][y]))) or np.array_equal(np.full((3), i % 2), np.array((arr[0][0], arr[1][1], arr[2][2]))) or np.array_equal(np.full((3), i % 2), np.array((arr[2][0], arr[1][1], arr[0][2]))):
                show_gril(arr)
                return f'Player {(i % 2)+1} Won'
        if not 2 in arr.reshape(9):
            return 'Draw'
        show_gril(arr)
```

```python
morpion()
```
