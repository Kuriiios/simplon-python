### 9

```python
categories = ['A', 'B', 'C', 'D']
valeurs1 = [10, 15, 20, 25]
valeurs2 = [5, 10, 15, 20]
fig, ax = plt.subplots()
bottom = np.zeros(4)
for values in [valeurs1, valeurs2]:
    p = ax.bar(categories, values, bottom=bottom)
    bottom += values
plt.show()
```
