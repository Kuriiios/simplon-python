```python
import numpy as np

answer = np.random.randint(1, 101)
n_try = 5
print(answer)

while n_try > 0:
    guess = int(input("Guess the number between 1 and 100"))
    match guess:
        case _ if guess > 100:
            int(input("Should be less or equal to 100"))
        case _ if guess < 1:
            int(input("Should be more or equal to 1"))

    match [guess > answer, guess == answer]:
        case [True, False]:
            print('Less')
        case [False, True]:
            print('Exact')
            break
        case [False, False]:
            print('More')
    n_try -= 1
            
```
