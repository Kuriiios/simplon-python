```python
PATH  = r'.venv' 
Nroots = sum(len(root) for root, dirnames, filesnames in os.walk(PATH) )
Ndirs = sum(len(dirnames) for root, dirnames, filesnames in os.walk(PATH) )
Nfiles = sum(len(filesnames) for root, dirnames, filesnames in os.walk(PATH) ) 
print(f"Number of roots: {Nroots}, Number of files: {Nfiles}, Number of directories: {Ndirs}")
```
