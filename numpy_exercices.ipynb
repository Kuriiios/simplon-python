### 9

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def is_prime_in_array(array):
    vec_func = np.vectorize(is_prime)
    mask = vec_func(array)
    return array[mask]

array =  np.array([np.random.randint(1, 20) for i in range(10)])
print(array)
is_prime_in_array(array)
```
